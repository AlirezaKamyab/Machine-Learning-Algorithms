# K-mean clustering
Here I will implement k-mean clustering algorithm used for clustering unlabeled data

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

The algorithm is as follows: <br>
```python
# Randomly initialize K cluster centeroids mu_1, mu_2, . . ., mu_K
Repeat:
    # Assign each point (data) to cluster centroids
    for i=1 to m:
        idx[i] = # index from 1 to k of cluster centroids which is closest to X[i]
    
    # Move cluster centroids
    for k=1 to K:
        centeroids[k] = #average (mean) of points assigned to cluster "k"
```

Distance of each of the points to the cluster centroids is calculated as such:<br>
$
    ||X[i] - mu||^{2}
$
<br>
The cost function will be the mean distance to each centroids given $\mu_c^{(i)}$ is the cluster centroid for $X^{(i)}$ <br>
$C^{(i)}$ is the index of cluster (1...K) which $X^{(i)}$ is assigned to.
<br>
$
    J(C^{1}\dots C^{m}, \mu_1 \dots \mu_K) = \displaystyle\sum_{i=1}^{m}||X^{(i)} - \mu_c^{(i)}||^2
$
<br>
The cost is called Distortion cost function

````python
def norm(x):
    return np.linalg.norm(x)


def calc_distances_to_centroids(X, idx, centroids):
    distances = np.zeros(X.shape[0])
    m,n = X.shape
    for i in range(m):
        distances[i] = norm(X[i] - centroids[idx[i]]) ** 2
    return distances


def cost(X, idx, centroids):
    summation = calc_distances_to_centroids(X, idx, centroids)
    summation = np.sum(summation)
    return summation / X.shape[0]
````

If we have m points, we should assign each point to the nearest cluster centroid (1...K). For each we find the distance to every centroid and we choose the minimum.

````python
def assign_centroids(X, centroids):
    idx = np.zeros(X.shape[0], dtype=int)
    for i in range(X.shape[0]):
        distances = np.zeros(centroids.shape[0])
        for k in range(centroids.shape[0]):
            distances[k] = norm(X[i] - centroids[k])
        idx[i] = np.argmin(distances)
    return idx
````

After each assignment we need to move centroids to the mean of their cluster.

````python
def move_centroids(X, idx, centroids):
    m, n = X.shape
    K = centroids.shape[0]
    for k in range(K):
        points = X[idx == k]
        centroids[k] = np.mean(points, axis=0)
    return centroids
````

And now we write the algorithm

````python
def k_mean(X, initial_centroids, max_iter):
    centroids = initial_centroids
    K = initial_centroids.shape[0]
    idx = np.zeros(X.shape[0], dtype='int')
    for i in range(1, max_iter + 1):
        print(f'Iteration {i}/{max_iter}')
        idx = assign_centroids(X, centroids)
        centroids = move_centroids(X, idx, centroids)
        print(f'Cost: {cost(X, idx, centroids)}')
    return centroids, idx
````

````python
def random_initial_centroids(X, K):
    a = np.random.permutation(X.shape[0])
    return X[a[:K]]
````

## Test

````python
X = np.array([[1, 1], [2, 2], [2, 6], [1, 4], [3, 6],
              [7, 5], [7, 8], [9, 6], [8, 7], [10, 8]], dtype=np.float64)
K = 2
initial_centroids = random_initial_centroids(X, K)
````

````python
plt.figure(figsize=(10, 5))
plt.scatter(X[:, 0], X[:, 1]);
plt.xlabel('$x_1$');
plt.ylabel('$x_2$');
````

````python
print(f'initial centroids are\n {initial_centroids}')
centroids, idx = k_mean(X, initial_centroids, 10)
````

````python
plt.figure(figsize=(10, 5))
plt.scatter(X[idx == 0][:, 0], X[idx == 0][:, 1], c='red', label='cluster 1');
plt.scatter(X[idx == 1][:, 0], X[idx == 1][:, 1], c='green', label='cluter 2');
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', marker='X', linewidths=1, label='centroids');
plt.xlabel('$x_1$');
plt.ylabel('$x_2$');
plt.legend();
````

````python
import cv2
import seaborn as sns
X = cv2.imread('./bird.jpeg')
````

````python
w, h, c = X.shape
````

````python
plt.imshow(X);
````

````python
X_data = X.reshape(w * h, -1)
X_data.shape
````

````python
X_data = X_data / 255
````
